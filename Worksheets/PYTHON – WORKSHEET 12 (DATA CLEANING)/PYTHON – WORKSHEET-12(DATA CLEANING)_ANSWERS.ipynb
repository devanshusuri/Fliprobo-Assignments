**7. Determine the row indexes having missing values.**

**9. Determine the column names having more than 30% missing values and remove them from the dataset.**

```python
pc_null = df.isnull().sum() / len(df)
missing_features = pc_null[pc_null > 0.30].index
print("\nThe column  having more than 30% missing values are \n:",missing_features)
df.drop(missing_features, axis=1, inplace=True)
print("\nAfter remove more than 30% missing value columns remaining columns left are \n:",df.columns)
df.shape
```

**10. Determine the number and percentage of rows having more than 5 missing values. If the percentage of such
rows is less than 20% then remove all these rows**

```python
d = df.loc[df.isnull().sum(axis=1)>5]
a=len(d)
b=round(len(d)/df.shape[0] * 100,2)
print(f'number and percentage of rows having more than 5 missing values are {a} and  {b}%')
```

```python
# percentage of such rows is less than 20% so remove all these rows
df.drop(df.loc[df.isnull().sum(axis=1)>5].index,axis=0,inplace=True)
df.shape
```

**12. Considering the 'Price' column as target variable which is to be predicted, how will you treat the missing values?
Give a brief explanation and treat the missing values in 'Price' column**

We have treat the missings values in price column with mean() of its values because if we drop all nan values in price column we lost large amount of information so we have three opton to replace nan values with mean,median and mode as distribution for the price column data is symmetric in nature so i'll use mean().

```python
#replacing missing values with mean in price
df['Price'].fillna(df['Price'].mean(), inplace=True)
```

**13. Now check the percentage of missing values in each column again, if missing values exist, treat it with the most
suitable method.**

```python
percent_missing = round((df.isnull().sum()*100)/len(df),2)

print("\nTotal number of empty rows percentage in each column:\n", percent_missing,"\n\n")
```

```python
#Longtitude and Lattitude contains very less nan values so we can drop these nan values
df = df[df['Longtitude'].notna()]
df = df[df['Lattitude'].notna()]
#replacing missing values with mean in Landsize
df['Landsize'].fillna(df['Landsize'].mean(), inplace=True)
#replacing missing values with mode in Car
df['Car'].fillna(df['Car'].mode()[0], inplace=True)
#replacing missing values with mode in Bedroom2
df['Bedroom2'].fillna(df['Bedroom2'].mode()[0], inplace=True)
#replacing missing values with mode in Bathroom
df['Bathroom'].fillna(df['Bathroom'].mode()[0], inplace=True)
```

```python
# Lets check again percentage of missing values after treating nan values in each field
percent_missing = round((df.isnull().sum()*100)/len(df),2)

print("\nTotal number of empty rows percentage in each column:\n", percent_missing,"\n\n")
```
