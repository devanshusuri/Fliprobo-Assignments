 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.


```python
import  requests
from  bs4 import BeautifulSoup

url = requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(url.text, 'html.parser')
story = soup.find_all(['h1', 'h2','h3'])
for i in story:
    print(i)
```

2. Write a python program to display IMDB’s Top rated 100 movies’ data(i.e. Name, IMDB rating, duration,
director, star cast - 10 main characters)

```python
from selenium import webdriver
url="https://www.imdb.com/chart/top/?ref_=nv_mv_250"
# keyword='Data Science'
driver=webdriver.Chrome(executable_path='chromedriver')
driver.get(url) # this command will search the given url in the browser
title=[]
rating=[]
director=[]
duration=[]
cast_name=[]
data={}
for i in range(1,101):
    driver.implicitly_wait(5)
    element=driver.find_element_by_xpath(f"//tbody[@class='lister-list']//tr[{i}]//td[@class='titleColumn']//a")
    element.click()
    ttl=driver.find_element_by_xpath("//div[@class='title_wrapper']//h1").text
    title.append(ttl)
    
    rat=driver.find_element_by_xpath("//div[@class='ratingValue']//span").text
    rating.append(rat)
    
    dirct=driver.find_element_by_xpath("//div[@class='credit_summary_item']//a").text
    director.append(dirct)
    
    dur=driver.find_element_by_xpath("//time").text
    duration.append(dur)
    
    try:
        casts=[]
        for j in range(2,11):
            cast=driver.find_element_by_xpath(f"//table[@class='cast_list']//tr[{j}]").text
            casts.append(cast)
    except:
        pass
    cast_name.append(casts)  
    driver.back()
    data={'Movie Name':title,'IMDB Rating':rating,'Director':director,'Duration':duration,'Star Cast':cast_name}

import pandas as pd
df_imdb=pd.DataFrame(data)
df_imdb
```

4. Write a python program to scrap book name, author name, genre and book review of 5 books from
‘www.bookpage.com’

```python
from selenium import webdriver
url="https://bookpage.com/reviews"
# keyword='Data Science'
driver=webdriver.Chrome(executable_path='chromedriver')
driver.get(url)
book_details={}
author=[]
reviews=[]
genre=[]
book_name=[]

# driver.find_element_by_xpath("//ul[@class='nav bp-navtext']//li[2]//a").click()
# driver.implicitly_wait(5)
# k=driver.find_element_by_xpath("//ul[@class='nav bp-navtext']//li[2]//ul//li//a").click()
for i in range(1,6):
    review=[]
    name=driver.find_element_by_xpath(f"//div[@class='row-fluid article-row'][{i}]//h4//a").text
    book_name.append(name)
    
    auth=driver.find_element_by_xpath(f"//div[@class='row-fluid article-row'][{i}]//p").text
    author.append(auth)
    
    gen=driver.find_element_by_xpath(f"//div[@class='row-fluid article-row'][{i}]//p[2]//a").text
    genre.append(gen)
    
    driver.find_element_by_xpath(f"//div[@class='row-fluid article-row'][{i}]//div[@class='read-full']//a").click() 
    driver.implicitly_wait(5)
    for j in range(2,5):
        rev=driver.find_element_by_xpath(f"//div[@class='article-body']//p[{j}]").text
        review.append(rev)
    driver.back()
    reviews.append(review)
book_details={"Book Name":book_name,"Author Name":author,"Genre":genre,"Review":reviews}

import pandas as pd
df_book=pd.DataFrame(book_details)
df_book
```
