#### 14. Write a python program to scrap 10 images of Rayban Sunglasses from flipkart website and save them in a directory.

```python
# Importing Useful libraries
import selenium
import os
import requests
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
# saving images in thw directory..
def save_images(data, dirname):
    for index,link in enumerate(data):
        print(f"Downloading {index+1} of {len(data)} images ")
        response=requests.get(link)
        with open(f"{dirname}/Rayban_Image_{index+1}.jpeg","wb") as file:
            file.write(response.content)

# Lets  make a directory "Flipkart_Image" to store the scrapped images..
data_path=os.getcwd()
path=os.path.join(data_path,'Flipkart_Images')
if not os.path.exists(path):
    os.makedirs(path)

    
# Making a functyion which will open flipkart website and then enter the rayban Glass keyword and then,
# crawl through the obtained results and stored the images link in a list...
def Scraping_images():
    driver=webdriver.Chrome(executable_path="chromedriver")
    url="https://www.flipkart.com/"
    driver.get(url)

    driver.find_element_by_xpath("//button[@class='_2AkmmA _29YdH8']").click()
    element=driver.find_element_by_xpath("//div[@class='O8ZS_U']//input")
    element.click()
    driver.implicitly_wait(5)
    element.send_keys("Rayban Glasses")
    element.send_keys("\n")
    images=[]
    k=1
    for i in range(1,11):
        for j in range(1,5):
            if len(images)<=9:
                img=driver.find_element_by_xpath(f"//div[@class='bhgxx2 col-12-12'][{i}]//div[@style='width: 25%;'][{j}]//div[@class='IIdQZO _1SSAGr']//a//img")
                images.append(img.get_attribute("src"))
                print(f"Scrapping Rayban glasses image {k}")
                k+=1
            else:
                break      
    return images



# Scraping image links
scrap_images=Scraping_images()
# Downloading and Saving Images in  the Directory
save_images(scrap_images,'Flipkart_Images')
# Displaying downloaded images from the directory..
filenames = os.listdir("Flipkart_Images")
filenames
```

#### 15. Write a program to scrap 20 mouses’ data which includes model name, price from www.amazon.in whose priceis less than 500 and make a data frame with 2 columns “model_name”,” price” with the scraped data

```python
# Importing Useful libraries
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException

# Setting up chrome as the automated webdriver
driver=webdriver.Chrome(executable_path="chromedriver")
url="https://www.amazon.in/"
driver.get(url)

# This will access the search bar
element=driver.find_element_by_xpath("//div[@id='nav-search']//div[@class='nav-search-field ']//input")
element.click()

# This will enter the required keyword we want to search on amazon website...
element.send_keys("Mouse")
element.send_keys("\n")

# defining Empty lists to store name and price information
names=[]
prices=[]


for j in range(1,20):
    for i in range(1,23):
        # model name
        if len(names)<20:
            driver.implicitly_wait(10)
            name=driver.find_element_by_xpath(f"//div[@data-component-type='s-search-result'][{i}]//h2//span").text

            # price
            try:
                price=driver.find_element_by_xpath(f"//div[@data-component-type='s-search-result'][{i}]//div[@class='sg-col-4-of-12 sg-col-8-of-16 sg-col-16-of-24 sg-col-12-of-20 sg-col-24-of-32 sg-col sg-col-28-of-36 sg-col-20-of-28']//span[@class='a-price']//span//span[2]").text
                price=float(price)
                if price < 500:
                    names.append(name)
                    prices.append(price)
                    
            except NoSuchElementException:
                continue  
        else:
            break
    # navigating to next page
    driver.find_element_by_xpath("//ul[@class='a-pagination']//li[@class='a-last']//a").click()
print("Scrapping of Mouse data with Price less than 500 is Completed")
print ("*" * 100)
print("Number of products Scrapped: ",len(names))
print ("*" * 100)

# Creating a DataFrame  to store the Model_name of the mouse and their respective price ,scraped from amazon
df=pd.DataFrame(names,columns=['Model_Names'])
df['Price(Rupees)']=prices
df
```
